```python
from google.colab import drive
drive.mount('/content/drive')
import os
import os.path as osp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DISTRIBUTION_MODES = ["balanced", "unbalanced_label", "unbalanced_labels_size"]
DATA_MODES = ["R", "S", "S1000"]
METHODS = ["sst_uka", "sst_ukk", "sst_ukl", "sst_umg", "sst_imise", "sst_hsm", "iil", "ciil"]

BASE_DIR = "/content/drive/MyDrive/DSM-AS"
DATA_DIR = osp.join(BASE_DIR, "input")
RESULT_DIR = osp.join(BASE_DIR, "output")
results = {(data_mode, distribution_mode, method): pd.read_csv(osp.join(RESULT_DIR, data_mode, distribution_mode, f"{method}_result.csv")) for data_mode in DATA_MODES for distribution_mode in DISTRIBUTION_MODES for method in METHODS}
```

```python
for distribution_mode in DISTRIBUTION_MODES:
    print("##################################################################################")
    print("Start distribution mode: {}".format(distribution_mode))
    print("##################################################################################")
    for method in METHODS:
        val_accs = []
        val_precs = []
        val_recs = []
        val_f1s = []
        for data_mode in DATA_MODES:
            df = results[data_mode, distribution_mode, method]
            val_accs.append(df.iloc[-1]["val_acc"])
            val_precs.append(df.iloc[-1]["val_prec"])
            val_recs.append(df.iloc[-1]["val_rec"])
            val_f1s.append(df.iloc[-1]["val_f1"])
        print(method)
        print("accuracy {:.1f}/{:.1f}/{:.1f} precsion {:.1f}/{:.1f}/{:.1f} recall {:.1f}/{:.1f}/{:.1f} f1-score {:.1f}/{:.1f}/{:.1f}".format(val_accs[0]*100, val_accs[1]*100, val_accs[2]*100, val_precs[0]*100, val_precs[1]*100, val_precs[2]*100, val_recs[0]*100, val_recs[1]*100, val_recs[2]*100, val_f1s[0]*100, val_f1s[1]*100, val_f1s[2]*100))
```
